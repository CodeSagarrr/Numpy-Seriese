# Start with basics of numpy with SAGAR 

```python
import numpy as np
import time
```

# Creating array in numpy

```python
arr_1d = np.array([1,2,3,4])
print("1D array : " ,arr_1d)

arr_2d = np.array([[1,2,3] , [4,5,6]])
print("2D array : " ,arr_2d)
```

# List vs Array

```python
py_list = [1,2,3,4] # Add another list to in the list
print("List by multiply 2 :" , py_list * 2)

np_array = np.array([1,2,3,4]) # Element wise multiplication
print("Numpy array by multiply 2 :" , np_array * 2)

# Time taken by list and numpy array
start = time.time()
```
