# Creating array in numpy

```python
arr_1d = np.array([1,2,3,4])
print("1D array : " ,arr_1d)

arr_2d = np.array([[1,2,3] , [4,5,6]])
print("2D array : " ,arr_2d)
```

### List vs Numpy array

```python
py_list = [1,2,3,4] # Add another list to in the list
print("List by multiply 2 :" , py_list * 2)

# Element wise multiplication
numpy_arr = np.array([1,2,3,4])
print("Numpy array by multiply 2 :" , numpy_arr * 2)
```

# List vs Numpy array who is efficient

```python


start = time.time()
py_list = [i * 2 for i in range(1000000)]
print("Time taken by manual list in python : " , time.time() - start)

start = time.time()
np_arr = np.arange(1000000) * 2
print("Time taken by numpy list in python : " , time.time() - start)
```

# Array creation in numpy 

```python
# 1 : Zero
zero_arr = np.zeros((4 , 4))
print("This is zero array zreation method : \n" , zero_arr)

# 2 : ones
one_arr = np.ones((4 , 4))
print("This is one array zreation method : \n" , one_arr)
```

```python
# Other array method for creation 
#  We have a full is defing constant fillups value for rows and colums , option value is constant in parameter 

full = np.full((3 , 4) , 6)  
print("Fill method : \n" , full)

#  random method which is mostly used in numpy it fill value randomly

random = np.random.random((4 , 4)) * 10
print("random method : \n" , random)

#  arange method print value in sequence , it takes optional parameter whhich is for steps how would you like to increase
sequence = np.arange(0 , 20 , 4) 
print("sequence method : \n" , sequence)
```

<!-- Now vector , metrices , tensor -->

```python
# Vector it is a simple array
vector = np.array([1, 2, 3, 4, 5])
print("Vector :" , vector)


# Metrices it is a combination of rows and columns in array
metrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("Metrix :" , metrix)

# Tensor it is besically multidimentional array like in one metrix we have other two metrix
tensor = np.array([[[1, 2] , [3, 4]],
                  [[5, 6] , [7, 8]]])
print("Tensor :" , metrix)
```

```python
#  numpy array properties ther are lots of properies bu we will explore important methods

metrix = np.array([[1, 2, 3],
                   [4, 5, 6]])
print("Metrix :" , metrix)
 #  1 : shape
print("Shape  : " , metrix.shape)

 #  2 : Dimention
print("Dimention  : " , metrix.ndim)

 #  3 : size
print("Size  : " , metrix.size)

 #  1 : Data type
print("Data type  : " , metrix.dtype)
```

```python
#  Array Reshape
arr = np.arange(10)
print("Sequence array :" , arr)

#  Array Reshape method it take rows and column parameter and conver sequence to metrix
reshape = arr.reshape((2,5))
print("reshape array :" , reshape)

#  Flatten method it convert metrix to seqnence array , and it print copy of reshape metrix
flatten = reshape.flatten()
print("flatten array :" , flatten)

# Ravel method also work like flatten but it print real value in sequence metrix instead of copy vale
ravel = reshape.ravel()
print("ravel array :" , ravel)

#  Transpose method convert order of metrix rows to colums and columns to rows
transpose = reshape.T
print("transpose array :" , transpose)
```
